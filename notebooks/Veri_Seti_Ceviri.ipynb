```python
# Gerekli k√ºt√ºphaneleri y√ºkle
!pip install deep-translator --quiet
import pandas as pd
from deep_translator import GoogleTranslator
from google.colab import files
import time

# üìÅ CSV dosyasƒ±nƒ± y√ºkle
uploaded = files.upload()

# üìñ Dosyayƒ± oku
df = pd.read_csv("balanced_fake_reviews_5000.csv")

# üîÅ ƒ∞ngilizce'den T√ºrk√ße'ye √ßeviri i≈ülemi
translated = []
for i, text in enumerate(df["text_"]):
    try:
        translated_text = GoogleTranslator(source='en', target='tr').translate(text)
    except Exception as e:
        translated_text = "√áEVƒ∞Rƒ∞ HATASI"
    translated.append(translated_text)

    if i % 100 == 0:
        print(f"{i} yorum √ßevrildi...")
        time.sleep(1)  # Google Translate'e √ßok hƒ±zlƒ± gitmemek i√ßin

# üî§ Yeni s√ºtunu ekle
df["translated_text"] = translated

# üíæ Yeni CSV olarak kaydet
df.to_csv("translated_reviews_5000.csv", index=False)

# üíª Dosyayƒ± indir
files.download("translated_reviews_5000.csv")
```
