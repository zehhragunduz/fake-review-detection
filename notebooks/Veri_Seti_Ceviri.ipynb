```python
# Gerekli kütüphaneleri yükle
!pip install deep-translator --quiet
import pandas as pd
from deep_translator import GoogleTranslator
from google.colab import files
import time

# 📁 CSV dosyasını yükle
uploaded = files.upload()

# 📖 Dosyayı oku
df = pd.read_csv("balanced_fake_reviews_5000.csv")

# 🔁 İngilizce'den Türkçe'ye çeviri işlemi
translated = []
for i, text in enumerate(df["text_"]):
    try:
        translated_text = GoogleTranslator(source='en', target='tr').translate(text)
    except Exception as e:
        translated_text = "ÇEVİRİ HATASI"
    translated.append(translated_text)

    if i % 100 == 0:
        print(f"{i} yorum çevrildi...")
        time.sleep(1)  # Google Translate'e çok hızlı gitmemek için

# 🔤 Yeni sütunu ekle
df["translated_text"] = translated

# 💾 Yeni CSV olarak kaydet
df.to_csv("translated_reviews_5000.csv", index=False)

# 💻 Dosyayı indir
files.download("translated_reviews_5000.csv")
```
