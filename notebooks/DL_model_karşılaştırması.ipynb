```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Veriler
data = {
    "Model": ["BERTurk", "BiLSTM", "RNN"],
    "Accuracy": [0.92, 0.82, 0.80],
    "F1-Score (Macro Avg)": [0.92, 0.81, 0.81]
}
df = pd.DataFrame(data)

# Sayısal konumlar
x = np.arange(len(df["Model"]))
width = 0.30  # Çubuk kalınlığı azaltıldı

# Grafik
plt.figure(figsize=(10, 5))
plt.bar(x - width/2, df["Accuracy"], width=width, label="Accuracy")
plt.bar(x + width/2, df["F1-Score (Macro Avg)"], width=width, label="F1-Score (Macro Avg)")

plt.xticks(x, df["Model"])
plt.ylim(0.7, 0.95)
plt.title("Model Karşılaştırması (Accuracy vs F1-score)")
plt.ylabel("Skor")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
